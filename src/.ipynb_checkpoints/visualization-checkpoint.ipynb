## Visualization

Read the resulting text file, and display a 2d plot

```python
f = open("results/u_10/part-00000", "r")
lines = f.readlines()
length = len(lines)
f.close()

x = []
y = []

for i, line in enumerate(lines):
    line = line.replace("[", "")
    line = line.replace("]", "")
    x.append(line.split(",")[0])
    y.append(line.split(",")[1])
```

```python
f = open("results/info_10/part-00000", "r")
lines = f.readlines()
length = len(lines)
f.close()

names = []
parties = []
sexes = []
constituencies = []
years = []

for i, line in enumerate(lines):
    line = line.replace("[", "")
    line = line.replace("]", "")
    names.append(line.split(",")[0])
    parties.append(line.split(",")[1])
    sexes.append(line.split(",")[2])
    constituencies.append(line.split(",")[3])
    years.append(line.split(",")[4].replace("\n", ""))
```

```python
parties
```

```python
symbols = []
for i in parties:
    if i == "M":
        symbols.append("bo")
    elif i == "S":
        symbols.append("ro")
    elif i == "KD":
        symbols.append("b+")
    elif i == "MP":
        symbols.append("go")
    elif i == "SD":
        symbols.append("yo")
    elif i == "C":
        symbols.append("g+")
    elif i == "V":
        symbols.append("r+")
    elif i == "L":
        symbols.append("co")
    else:
        symbols.append("ko")
        
print(len(parties))
print(len(symbols))
        
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
for i in range(0, len(x)):
    plt.plot(x[i], y[i], symbols[i])
```

```python
symbols = []
for i in sexes:
    if i == "man":
        symbols.append("bo")
    elif i == "kvinna":
        symbols.append("ro")
        
print(len(parties))
print(len(symbols))
        
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
for i in range(0, len(x)):
    plt.plot(x[i], y[i], symbols[i])
```

```python
years
```

```python
print(min(years))
print(max(years))

symbols = []
for i in years:
    if int(i) <= 1938:
        symbols.append("ko")
    elif int(i) > 1938 and int(i) <= 1948:
        symbols.append("bo")
    elif int(i) > 1948 and int(i) <= 1958:
        symbols.append("co")
    elif int(i) > 1958 and int(i) <= 1968:
        symbols.append("go")
    elif int(i) > 1968 and int(i) <= 1978:
        symbols.append("yo")
    elif int(i) > 1978 and int(i) <= 1988:
        symbols.append("ro")
    else:
        symbols.append("mo")
        
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
for i in range(0, len(x)):
    plt.plot(x[i], y[i], symbols[i])
```
